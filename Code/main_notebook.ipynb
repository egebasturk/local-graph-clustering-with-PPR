Merge Clusters according the result of Following Formula
$$ {C_i \cap C_j}\over{\min{(C_i, C_j)}}$$


```python
cluster_merge_threshold = 0.85

tmp = len(prS)
i = 0
while i < tmp:
    p = i
    while p < tmp:
        if (len(prS[i].keys() & prS[p].keys())
                / min(len(prS[i]), len(prS[p])) > cluster_merge_threshold):
            prS[i].update(prS[p])
            del prS[p]
            tmp = tmp - 1
        p = p + 1
    i = i + 1

print(prS[0])
print(len(prS))
```
